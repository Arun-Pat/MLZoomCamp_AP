Compute the inverse of XTX.
Create an array y with values [950, 1300, 800, 1000, 1300].
Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.

```python
df['ocean_proximity'].unique()
```

```python
# Filter values in DF, then select columns, then convert DF to list
# X = df[df['ocean_proximity']=='ISLAND'][['housing_median_age', 'total_rooms', 'total_bedrooms']].values.tolist()
# Filter values in DF, then select columns, then convert DF to NupPy array
X = df[df['ocean_proximity']=='ISLAND'][['housing_median_age', 'total_rooms', 'total_bedrooms']].to_numpy()
```

```python
X
```

```python
#XTX = X.transpose() # Transpose of X
XT= X.T # Transpose of X
XT
```

```python
XTX = XT.dot(X)
XTX
```

```python
XTX_Inv = np.linalg.inv(XTX)
```

```python
XTX_Inv
```

```python
y=[950, 1300, 800, 1000, 1300]
y
```

```python
w = XTX_Inv.dot(XT)
w.dot(y)
```
